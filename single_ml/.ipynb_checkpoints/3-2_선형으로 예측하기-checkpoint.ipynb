```python


import numpy as np
perchLength = np.array(
    [8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0,
     21.0, 21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5,
     22.5, 22.7, 23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5,
     27.3, 27.5, 27.5, 27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0,
     36.5, 36.0, 37.0, 37.0, 39.0, 39.0, 39.0, 40.0, 40.0, 40.0,
     40.0, 42.0, 43.0, 43.0, 43.5, 44.0]
     )
perchWeight = np.array(
    [5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0,
     110.0, 115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0,
     130.0, 150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0,
     197.0, 218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0,
     514.0, 556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0,
     820.0, 850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0,
     1000.0, 1000.0]
     )
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainInput, testInput, trainTarget, testTarget = train_test_split(perchLength, perchWeight, random_state=42)
trainInput = trainInput.reshape(-1, 1)
testInput = testInput.reshape(-1,1)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor(n_neighbors=3)
knr.fit(trainInput, trainTarget)

knr.predict([[50]])
```

```python
import matplotlib.pyplot as plt

distances, indexs = knr.kneighbors([[50]])
plt.scatter(trainInput, trainTarget)
plt.scatter(trainInput[indexs], trainTarget[indexs], marker='d')
plt.scatter(50, 1033, marker='^')
plt.xlabel('len')
plt.ylabel('wei')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(trainInput, trainTarget)
lr.predict([[50]])
```

```python
#모델 가중치 확인 
print(lr.coef_, lr.intercept_)
```

```python
plt.scatter(trainInput, trainTarget)

plt.plot([10, 45], [10*lr.coef_+lr.intercept_, 45*lr.coef_+lr.intercept_])


plt.scatter(50, 1241.8, marker='^')
plt.xlabel('len')
plt.ylabel('wei')
plt.show()
```

```python
plt.scatter(trainInput, trainTarget)

plt.plot([15, 50], [15*lr.coef_+lr.intercept_, 50*lr.coef_+lr.intercept_])

plt.scatter(50, 1241.8, marker='^')
plt.xlabel('len')
plt.ylabel('wei')
plt.show()
```

```python
print(lr.score(trainInput, trainTarget))
print(lr.score(testInput, testTarget))
    
```

```python
lr = LinearRegression()
lr.fit(trainPoly, trainTarget)


print(lr.predict([[50**2, 50]]))
print(lr.coef_, lr.intercept_)
```
